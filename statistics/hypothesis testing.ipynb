```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import math
```

```python
mu=250
sigma=12
n=40
sample_mean=246.5

standerd_error=sigma/math.sqrt(n)
z_score=(sample_mean-mu)/standerd_error

p_value_manual=2*(1-stats.norm.cdf(abs(z_score)))

print(f'--manualk calculation---')
print(f'z-statistics: {z_score:.4f}')
print(f'p_value:{p_value_manual:.4f}') # if pvalue > critical than reject 

if p_value_manual< 0.05:
    print('reject the null hypothesis')
else:
    print('accept the null hypothesis')
```
