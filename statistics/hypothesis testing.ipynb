```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import math
```

```python
mu=250
sigma=12
n=40
sample_mean=246.5

standerd_error=sigma/math.sqrt(n)
z_score=(sample_mean-mu)/standerd_error

p_value_manual=2*(1-stats.norm.cdf(abs(z_score)))

print(f'--manualk calculation---')
print(f'z-statistics: {z_score:.4f}')
print(f'p_value:{p_value_manual:.4f}') # if pvalue > critical than reject 

if p_value_manual< 0.05:
    print('reject the null hypothesis')
else:
    print('accept the null hypothesis')
```

```python
center_a=np.random.normal(75,10,120)
center_b=np.random.normal(78,12,150)
```

```python
mean_a,mean_b=np.mean(center_a), np.mean(center_b)
s1,s2=10,12
n1,n2=len(center_a),len(center_b)

z_2samp=(mean_a - mean_b)/ np.sqrt((s1 ** 2/n1) + s2 **2/n2)
p_2samp=stats.norm.sf(abs(z_2samp)) *2
print(z_2samp)
print(p_2samp)
if p_2samp <0.05:
    print('reject null hypothesis')
else:
    print('accept null hypothesis')
```

```python
z_stat_2,p_value_2=ztest(center_a,center_b)
print(f'two-sample z-test:{z_stat_2:.4f},p_value_2:{p_value_2:.4f}')

if p_value_2 <0.05:
    print('reject null hypothesis')
else:
    print('accepth null hypothesis')
```

```python
group_a=[20,22,19,21,23]
group_b=[25,28,24,26,27]

mean1,mean2=np.mean(group_a), np.mean(group_b)
var1,var2=np.var(group_a,ddof=1),np.var(group_b,ddof=1)
n1,n2= len(group_a),len(group_b)

pooled_se =math.sqrt((var1/n1) + (var2/n2))
t_manual= (mean1-mean2)/pooled_se

df=n1+n2 -2

p_manual=stats.t.sf(abs(t_manual),df)*2
print(f'manual test:{t_manual:.4f},manual p_value:{p_manual:.4f}')

if p_manual <0.05:
    
    print('reject null hypothesis')
else:
    print('accepth nulll hypothesis')
```

# chi squre test

```python
import numpy as np
import pandas as pd
import scipy.stats as stats
```

```python
observed=np.array([[150,100,50],
                   [100,150,150]])

row_totals=observed.sum(axis=1)
col_totals=observed.sum(axis=0)

grad_total=observed.sum()
expected=np.outer(row_totals,col_totals)/grad_total

chi_s_manual=np.sum((observed-expected)**2/expected)
df_manual=(observed.shape[0]-1)* (observed.shape[1]-1)

p_value=stats.chi2.sf(chi_s_manual,df_manual)
print('-- manual--')
print(f'chi_s_manual:{chi_s_manual:.4f}')
print(f'p_value:{p_value:.4f}')

if p_value<0.05:
    print('reject null hypothesis')
else:
    print('accept null hypothesis')
```
