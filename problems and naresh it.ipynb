```python
operation = """
1. cm to ft conversion
2. km to miles conversion
3. usd to inr conversion
4. exit
"""

while True:
    print(operation)
    user_choice = int(input("Enter your choice (1-4): "))

    if user_choice == 1:
        user = float(input("Enter a cm number: "))
        ft = user / 30.48
        print("Feet =", ft)

    elif user_choice == 2:
        user = float(input("Enter a km number: "))
        miles = user * 0.621371  
        print("Miles =", miles)

    elif user_choice == 3:
        user = float(input("Enter a USD amount: "))
        inr = user * 83 
        print("INR =", inr)

    elif user_choice == 4:
        print("Exiting program...")
        break

    else:
        print("Invalid choice, please try again.")
```
