# stemming --stem is give root node of the word

```python
from nltk.stem import PorterStemmer
pst=PorterStemmer()
```

```python
pst.stem('affection')
```

```python
pst.stem('playing')
```

```python
pst.stem('maximum')
```

```python
words=['give','giving','given','gave','git','thinking','nilesh','lovenig']
for word in words:
    print(word+ ' : ' + pst.stem(word))
```

# type of stemming-porterstremmer -remove the suffix not all the words
# lancasterstremmer - remove for every word

```python
from nltk.stem import LancasterStemmer
st=LancasterStemmer()

for word in words:
    print(word+ ': ' + pst.stem(word))
```

```python
from nltk.stem import SnowballStemmer
st=SnowballStemmer('german') # in this pass the language

for word in words:
    print(word+ ': ' + pst.stem(word))
```
