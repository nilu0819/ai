# word teckenize

```python
from nltk.tokenize import word_tokenize
```

```python
AI_tokens=word_tokenize(AI)
AI_tokens
```

```python
len(AI_tokens)
```

```python
from nltk.tokenize import sent_tokenize
AI_sent=sent_tokenize(AI)
```
