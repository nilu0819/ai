# array manupilation and functions

# (1).reshape()

```python
# changing the shape of the array withut changing the data
import numpy as np
a=np.arange(6)
print(a)

b=a.reshape((2,3))
print(b)
```

# (2).ravel()

```python
# flat the array into 1d array abd return= aview when a possible
print(b.ravel())
```

# (3).flatten()

```python
# is similer to ravel() but alwayes gives the return a copy
print(b.flatten())
```

# (4).transpose()

```python
#Swaps rows and columns (generalized for higher dimensions).
print(b.transpose())
```

# (5).swapaxes()

```python
# swap two axex in array 
c = np.arange(8).reshape(2,2,2)
print(c.swapaxes(0, 2))
```

# (6).moveaxis()

```python
#Moves an axis to a new position.
d = np.ones((2, 3, 4))
print(np.moveaxis(d, 0, -1).shape)  # (3, 4, 2)
```

# (7).concatenate()

```python
#Joins arrays along an existing axis
x = np.array([[1, 2], [3, 4]])
y = np.array([[5, 6]])
print(np.concatenate((x, y), axis=0))
```

# (8).stack()

```python
# Joins arrays along a new axis.
a = np.array([1, 2])
b = np.array([3, 4])
print(np.stack((a, b)))
```

# (9).hstack() and .vstack()

```python
#Horizontal and vertical stacking.
print(np.hstack((a, b)))  
print(np.vstack((a, b)))
```

# (10).split()

```python
#  splits an array into multiple sub-arrays.
arr = np.arange(9)
print(np.split(arr, 3))
```
