# classification problem

```python
import numpy as np
import pandas as pd
```

```python
np.random.seed(42)

n = 1000

data = pd.DataFrame({
    'offer': np.random.choice([0,1], size=n, p=[0.7,0.3]),
    'win': np.random.choice([0,1], size=n, p=[0.8,0.2]),
    'urgent': np.random.choice([0,1], size=n, p=[0.75,0.25])
})

data
```

```python
data['spam'] = (
    (data['offer'] == 1) &  ((data['win'] == 1) | (data['urgent'] == 1))
).astype(int)

data.head()
```

# calculate the prior margine probabilty

```python
# p (spam)
P_spam = data['spam'].mean() 
P_not_spam = 1 - P_spam

print("P(Spam):", P_spam)
print("P(Not Spam):", P_not_spam)
```

# calculate conditional probabilty
# p(offer = 1 | spam)

```python
P_offer_given_spam = len(data[(data.offer==1) & (data.spam==1)]) / len(data[data.spam==1])
P_offer_given_notspam = len(data[(data.offer==1) & (data.spam==0)]) / len(data[data.spam==0])

print(P_offer_given_spam)
print(P_offer_given_notspam)
```

# p(win=1| spam)

```python
P_win_given_spam=len(data[(data.win==1) & (data.spam==1)]) /len(data[data.spam==1])
P_win_given_notspam = len(data[(data.win==1) & (data.spam==0)]) / len(data[data.spam==0])

print(P_win_given_spam)
print(P_win_given_notspam)
```

# p(urgent =1 |spam)

```python
P_urgent_given_spam = len(data[(data.urgent==1) & (data.spam==1)]) / len(data[data.spam==1])
P_urgent_given_notspam = len(data[(data.urgent==1) & (data.spam==0)]) / len(data[data.spam==0])

print(P_urgent_given_spam)
print(P_urgent_given_notspam)
```

```python
data['spam'] = (
    (data['offer'] == 0) &  ((data['win'] == 1) | (data['urgent'] == 1))
).astype(int)

data.head()
```

# prior margineal probabillity

```python
P_spam = data['spam'].mean() 
P_not_spam = 1 - P_spam

print("P(Spam):", P_spam)
print("P(Not Spam):", P_not_spam)
```

# calculate the conditional probability
# offers =0| spam

```python
P_offer_given_spam = len(data[(data.offer==0) & (data.spam==1)]) / len(data[data.spam==1])
P_offer_given_notspam = len(data[(data.offer==0) & (data.spam==0)]) / len(data[data.spam==0])

print(P_offer_given_spam)
print(P_offer_given_notspam)
```

# win=1 | spam

```python
P_win_given_spam = len(data[(data.win==1) & (data.spam==1)]) / len(data[data.spam==1])
P_win_given_notspam = len(data[(data.win==1) & (data.spam==0)]) / len(data[data.spam==0])

print(P_win_given_spam)
print(P_win_given_notspam)
```

# urgent =1|spam

```python
P_urgent_given_spam = len(data[(data.urgent==1) & (data.spam==1)]) / len(data[data.spam==1])
P_urgent_given_notspam = len(data[(data.urgent==1) & (data.spam==0)]) / len(data[data.spam==0])

print(P_urgent_given_spam)
print(P_urgent_given_notspam)
```

```python
# Get needed probabilities

P_offer0_given_spam = 1 - (len(data[(data.offer==0) & (data.spam==1)]) / len(data[data.spam==1]))
# P_win0_given_spam = 1-0.48148148148148145

P_offer0_given_notspam = 1 - (len(data[(data.offer==0) & (data.spam==0)]) / len(data[data.spam==0]))

P_urgent_given_spam = len(data[(data.urgent==1) & (data.spam==1)]) / len(data[data.spam==1])

# Compute numerator for spam
numerator_spam = (
    P_spam *
    P_offer0_given_spam *
    P_win_given_spam *
    P_urgent_given_spam
)

print("Unnormalized Spam Probability:", numerator_spam)
```

```python
# Get needed probabilities

P_offer0_given_notspam = 1 - (len(data[(data.offer==0) & (data.spam==0)]) / len(data[data.spam==0]))
P_urgent_given_spam = len(data[(data.urgent==1) & (data.spam==1)]) / len(data[data.spam==1])

numerator_notspam = (
    P_not_spam *
    P_offer0_given_notspam *
    P_win_given_notspam *
    P_urgent_given_notspam
)
numerator_notspam
```

```python
print("Unnormalized Spam:", numerator_spam)
print("Unnormalized Not Spam:", numerator_notspam)
```

```python
if numerator_spam > numerator_notspam:
    print("Manual Prediction: SPAM")
else:
    print("Manual Prediction: NOT SPAM")
```
