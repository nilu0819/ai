```python
import numpy as np
```

# sasmple data

```python
np.random.seed(42)
sample=np.random.normal(loc=2000,scale=400,size=50)
sample
```

# point estimation

```python
mean=np.mean(sample)
std=np.std(sample,ddof=1) # delta degree of freedom
n=len(sample)

print(mean)
print(std)
print(n,"sample size")
```

# confidance interval for mean

# compute standerd error

```python
import scipy.stats as stats
```

```python
standerd_error=std / np.sqrt(n)
```

```python
standerd_error
```

# get the t -value with 95% confidance

```python
confidance=0.95
t_value=stats.t.ppf((1+confidance)/2,df=n-1) # ppf percent point function
t_value
```

# margine error

```python
margine_error= t_value * standerd_error
```

```python
margine_error
```

# confidence interval 

```python
lower= mean - margine_error
upper=mean + margine_error
print(lower,'and',upper)
```
