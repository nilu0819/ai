# keyword functionsa 

```python
def person(name,age):# in this fromal arguments 
    print(name)
    print(age-5)
    
person(age = 22, name = 'nit')# actual arguments with keyword
```

```python
def person(name, age):
    print(name)
    print(age)
    
person('michel')# mising the arguments
```

```python
def person(name, age=18):
    print(name)
    print(age)
    
person('michel')
```

```python
def person(name, age=18):
    print(name)
    print(age)
    
person('michel', 41)
```
