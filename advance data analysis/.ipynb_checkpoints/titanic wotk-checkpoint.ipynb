```python
def getnumber(str):
    if str =='male':
        return 1
    else:
        return 2
df['Gender']=df['Sex'].apply(getnumber)
df.head()
```

```python
del df['Sex']
df.head()
```

```python
df.isnull().sum()
```

```python
mean=df[df.Survived==1].Age.mean()
mean
```

```python
df["age"]=np.where(pd.isnull(df.Age) & df["Survived"]==1  ,mean, df["Age"])
df.head()
```

```python
df.isnull().sum()
```

```python
df.age.fillna(mean,inplace=True)
df.head()
```

```python
del df['Age']
df.head()
```

```python
survivedQ = df[df.Embarked == 'Q'][df.Survived == 1].shape[0]
survivedC = df[df.Embarked == 'C'][df.Survived == 1].shape[0]
survivedS = df[df.Embarked == 'S'][df.Survived == 1].shape[0]
print(survivedQ)
print(survivedC)
print(survivedS)
```

```python
survivedQ = df[df.Embarked == 'Q'][df.Survived == 0].shape[0]
survivedC = df[df.Embarked == 'C'][df.Survived == 0].shape[0]
survivedS = df[df.Embarked == 'S'][df.Survived == 0].shape[0]
print(survivedQ)
print(survivedC)
print(survivedS)
```

```python

df.dropna(inplace=True)
df.head()
```

```python
df.isnull().sum()
```

```python
df.rename(columns={'age':"Age"},inplace =True)
df.head()
```

```python
def getemb(str):
    if str=='S':
        return 1
    elif str=='Q':
        return 2
    else:
        return 3
df['Embarked']=df['Embarked'].apply(getemb)
df.head()
```

```python
import matplotlib.pyplot as plt
from matplotlib import style

males=(df['Gender']==1).sum()
females=(df['Gender']==2).sum()

print(males)
print(females)
p=[males,females]
plt.pie(p,
       labels=['Male','female'],
       colors=['red','blue'],
       explode=(0.15,0),
       startangle=0)
plt.axis('equal')
plt.show()
```

```python
males=df[df.Gender==1][df.Survived==1].shape[0]
print(males)
malen=df[df.Gender==1][df.Survived==0].shape[0]
print(malen)
FemaleS=df[df.Gender==2][df.Survived==1].shape[0]
print(FemaleS)
FemaleN=df[df.Gender==2][df.Survived==0].shape[0]
print(FemaleN)
```

```python
chart=[Males,Malen,FemaleS,FemaleN]
colors=['lightskyblue','yellowgreen','Yellow','Orange']
labels=["Survived Male","Not Survived Male","Survived Female","Not Survived Female"]
explode=[0,0.05,0,0.1]
plt.pie(chart,labels=labels,colors=colors,explode=explode,startangle=100,counterclock=False,autopct="%.2f%%")
plt.axis("equal")
plt.show()
```
