```python
import pandas as pd
import numpy as np

np.random.seed(42)

n = 1000

data = pd.DataFrame({
    'offer': np.random.choice([0,1], size=n, p=[0.7,0.3]),
    'win': np.random.choice([0,1], size=n, p=[0.8,0.2]),
    'urgent': np.random.choice([0,1], size=n, p=[0.75,0.25])
})

data
```

```python
# Define spam logic
data['spam'] = (
    (data['offer'] == 1) &  ((data['win'] == 1) | (data['urgent'] == 1))
).astype(int)

data.head()
```

```python
P_spam = data['spam'].mean() 
P_not_spam = 1 - P_spam

print("P(Spam):", P_spam)
print("P(Not Spam):", P_not_spam)
```

```python
P_offer_given_spam = len(data[(data.offer==1) & (data.spam==1)]) / len(data[data.spam==1])
P_offer_given_notspam = len(data[(data.offer==1) & (data.spam==0)]) / len(data[data.spam==0])

print(P_offer_given_spam)
print(P_offer_given_notspam)
```

```python
P_win_given_spam = len(data[(data.win==1) & (data.spam==1)]) / len(data[data.spam==1])
P_win_given_notspam = len(data[(data.win==1) & (data.spam==0)]) / len(data[data.spam==0])

print(P_win_given_spam)
print(P_win_given_notspam)
```

```python
P_urgent_given_spam = len(data[(data.urgent==1) & (data.spam==1)]) / len(data[data.spam==1])
P_urgent_given_notspam = len(data[(data.urgent==1) & (data.spam==0)]) / len(data[data.spam==0])

print(P_urgent_given_spam)
print(P_urgent_given_notspam)
```

```python
import pandas as pd
import numpy as np

np.random.seed(42)

n = 1000

data = pd.DataFrame({
    'offer': np.random.choice([0,1], size=n, p=[0.7,0.3]),
    'win': np.random.choice([0,1], size=n, p=[0.8,0.2]),
    'urgent': np.random.choice([0,1], size=n, p=[0.75,0.25])
})

# Define spam logic
data['spam'] = (
    (data['offer'] == 1) & ((data['win'] == 1) | (data['urgent'] == 1))
).astype(int)

# Probabilities
P_spam = data['spam'].mean()
P_not_spam = 1 - P_spam

P_offer_given_spam = len(data[(data.offer==1) & (data.spam==1)]) / len(data[data.spam==1])
P_offer_given_notspam = len(data[(data.offer==1) & (data.spam==0)]) / len(data[data.spam==0])

P_win_given_spam = len(data[(data.win==1) & (data.spam==1)]) / len(data[data.spam==1])
P_win_given_notspam = len(data[(data.win==1) & (data.spam==0)]) / len(data[data.spam==0])

P_urgent_given_spam = len(data[(data.urgent==1) & (data.spam==1)]) / len(data[data.spam==1])
P_urgent_given_notspam = len(data[(data.urgent==1) & (data.spam==0)]) / len(data[data.spam==0])

# Needed probabilities for case (offer=0, win=1, urgent=1)
P_offer0_given_spam = 1 - P_offer_given_spam
P_offer0_given_notspam = 1 - P_offer_given_notspam

# Numerators
numerator_spam = (
    P_spam *
    P_offer0_given_spam *
    P_win_given_spam *
    P_urgent_given_spam
)

numerator_notspam = (
    P_not_spam *
    P_offer0_given_notspam *
    P_win_given_notspam *
    P_urgent_given_notspam
)

print("Unnormalized Spam:", numerator_spam)
print("Unnormalized Not Spam:", numerator_notspam)

if numerator_spam > numerator_notspam:
    print("Manual Prediction: SPAM")
else:
    print("Manual Prediction: NOT SPAM")
```
