```python
df=pd.read_csv(r"C:\Users\niles\Downloads\placement.csv")
```

```python
plt.scatter(df['cgpa'],df['package'],color='red')
plt.grid()
plt.xlabel('cgpa')
plt.ylabel('package')
plt.show()
```

```python
x=df.iloc[:,0:1]
y=df.iloc[:,-1]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
x_test
```

```python
y_test
```

```python
lr.predict(x_test.iloc[10].values.reshape(1,1))
```

```python
plt.scatter(df['cgpa'],df['package'],color='pink')
plt.plot(x_train,lr.predict(x_train),c='blue')
plt.grid()
plt.xlabel('cgpa')
plt.ylabel('package')
plt.show()
```
